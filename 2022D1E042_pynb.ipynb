```python
import pandas as pd

df = pd.read_csv('genap.csv', encoding='latin-1')

print("JUMLAH DATA:", df.shape)
print(df.head())
```

```python
df['text'] = df.iloc[:, 0].fillna('').astype(str)

print(df[['text']].head())
```

```python
import re

def cleaning(text):
    text = re.sub(r'http\S+', ' ', text)
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip()
```

```python
df['cleaning'] = df['text'].apply(cleaning)
print(df[['text','cleaning']].head(10))
```

```python
def tokenizing(text):
    return text.split()


```

```python
df['tokenizing'] = df['cleaning'].apply(tokenizing)
```

```python
print(df[['cleaning','tokenizing']].head(10))
```

```python
normal_dict = {
    "tdk": "tidak",
    "yg": "yang",
    "dr": "dari",
    "utk": "untuk",
    "krn": "karena"
}
```

```python
def normalization(tokens):
    return [normal_dict.get(word, word) for word in tokens]
```

```python
df['normalization'] = df['tokenizing'].apply(normalization)
```

```python
print(df[['tokenizing','normalization']].head(10))
```

```python
from nltk.corpus import stopwords
import nltk

nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('indonesian'))
```

```python
def stopword_removal(tokens):
    return [word for word in tokens if word not in stop_words]
```

```python
df['stopword'] = df['normalization'].apply(stopword_removal)
```

```python
print(df[['normalization','stopword']].head(10))
```
